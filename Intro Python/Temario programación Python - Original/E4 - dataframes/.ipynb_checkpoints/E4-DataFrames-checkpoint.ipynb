# Data Frames

Uno de los objetivos de `pandas` es dar una funcionalidad de análisis de datos parecida a la proporcionada por `R`. Podemos encontrar una comparación en la documentación de pandas: https://pandas.pydata.org/pandas-docs/stable/comparison_with_r.html 

El `DataFrame` es el tipo de datos esencial de `pandas`. Tepresenta una estructura bidimensional de tabla: filas y columnas.

```python
import pandas
```

http://pandas.pydata.org/pandas-docs/stable/

Podemos construir un `DataFrame` de muchas formas. A partir de un fichero `cvs` es muy obvio, pues estos ficheros tienen ya esta estructura de tabla, con columnas que indican campos y filas que indican valores asociados a dichos campos.

```python
realestate = pandas.read_csv('realestate.csv')
realestate
```

```python
type(realestate)
```

```python
realestate.info()
```

Vamos a comentar a continuación cómo podemos acceder a los diferentes elementos de la tabla

### Acceso a columnas 

```python
realestate['price']
```

```python
realestate[['price','zip']]
```

```python
selcols = realestate.columns[2:6]
selcols
```

```python
realestate[selcols]
```

```python
mixed = list(selcols)+['price']
mixed
```

```python
realestate[mixed]
```

### Acceso a filas

```python
realestate.iloc[3]
```

```python
realestate.iloc[[3,6]]
```

```python
realestate.iloc[3:6]
```

```python
list(range(3,8,2))
```

```python
realestate.iloc[range(3,8,2)]
```

```python
realestate.iloc[3]['city']
```

```python
realestate.iloc[3][1]
```

```python
realestate[['price','zip']][3:8]
```

## Operaciones con Data Frame

El módulo `pandas` nos afrece una amplísima variedad de funciones para trabajar con los `DataFrame`, a continuación, a modo de ejemplo te mostramos algunas.

```python
realestate.sort_values(by='price', ascending=False)
```

```python
realestate.describe()
```

```python
cond = realestate['baths'] == 2
cond
```

```python
realestate[cond]
```

```python
two_baths = realestate['baths'] == 2
three_beds = realestate['beds'] == 3
realestate[two_baths & three_beds]
```

```python
realestate[two_baths & three_beds]['zip'].value_counts()
```

```python
realestate.groupby(['zip','baths']).size()
```

```python
realestate.groupby(['zip','baths']).size().unstack()
```
