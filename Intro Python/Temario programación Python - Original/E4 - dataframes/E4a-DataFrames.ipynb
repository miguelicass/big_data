Un dataframe es una estructura bidimensional, con filas y columnas, esto es, una tabla de datos.
El `DataFrame` es el tipo de datos esencial de la librería `pandas`. 

```python
import pandas
```

Podemos construir un `DataFrame` de muchas formas. Uno de los más empleados es a partir de un archivo `cvs`, que maneja esencialmente estas estructuras de tabla, con columnas que indican campos y filas que indican registros, cuyas celdas son los valores asociados a dichos campos.

```python
realestate = pandas.read_csv('realestate.csv')
realestate
```

```python
type(realestate)
```

```python
realestate.info()
```

## Operaciones con Dataframes

El módulo `pandas` nos afrece una amplísima variedad de funciones para trabajar con los `DataFrame`, a continuación, a modo de ejemplo te mostramos algunas.

```python
realestate.sort_values(by='price', ascending=False)
```

```python
realestate.describe()
```

```python
cond = realestate['baths'] == 2
cond
```

```python
realestate[cond]
```

```python
two_baths = realestate['baths'] == 2
three_beds = realestate['beds'] == 3
realestate[two_baths & three_beds]
```

```python
realestate[two_baths & three_beds]['zip'].value_counts()
```

```python
realestate.groupby(['zip','baths']).size()
```

```python
realestate.groupby(['zip','baths']).size().unstack()
```

## Matplotlib

```python
tabla_habs_precios = realestate[['beds','price']]
tabla_habs_precios
```

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

xs = tabla_habs_precios['beds']
ys = tabla_habs_precios['price']
slope, intercept, r_value, p_value, std_err = stats.linregress(xs, ys)
recta_regres = lambda x: intercept + slope*x

plt.plot(xs, ys, 'o', label='(hab, precio)')
plt.plot(xs, recta_regres(xs), 'r', label='recta de regresión')
plt.legend(loc = 'upper left')
plt.xlabel('número de dormitorios')
plt.ylabel('precio')
plt.show()
```
