```python
import numpy as np


def leer_datos(file):
    with open(file, 'r') as f:
        escanos = int(f.readline().strip())
        votos_str = f.readline().strip().split(" ")
        votos = [int(v) for v in votos_str]
        return escanos, votos
    
def reparto(escanos, votos):
    #votos_aux = votos * 1 
    #votos_aux = list(votos)
    votos_aux = votos[:]
    reparto_escanos = [0] * len(votos)
    #votos = [(v / (j + 1)) for v in votos]
    for i in range(escanos):
        print("Escaños: ", i )
        print(votos)
        print(votos_aux)
        max_votos = max(votos)
        index_max_votos = votos.index(max_votos)
        reparto_escanos[index_max_votos] = reparto_escanos[index_max_votos] + 1
        votos[index_max_votos] = votos_aux[index_max_votos] / (reparto_escanos[index_max_votos] + 1) 
        print(reparto_escanos)
        print("-----------------")
    return reparto_escanos


escanos, votos = leer_datos("./Datos_DHont_ej1.txt")
print(escanos, votos)
print("Reparto final de escaños:")
print(reparto(escanos, votos))
print("---------------------------------------------")
print("---------------------------------------------")
escanos, votos = leer_datos("./Datos_DHont_ej2.txt")
print(escanos, votos);
print("Reparto final de escaños:")
print(reparto(escanos, votos))
```

```python
def get_length(item):
    return len(item)

def trazar_diagrama_de_barras(reparto_escanos):
    keys = list(reparto_escanos.keys())
    keys.sort(key = get_length, reverse = True)
    clave_max_length = len(keys[0])
    for partido, num_escano in reparto_escanos.items():
        stars = "*" * num_escano
        space_length = " " * (clave_max_length - len(partido))
        print(partido, space_length, "|", stars)
            

trazar_diagrama_de_barras(
    {"Los Verdes": 9, "Izquierdas": 17, "Norteños": 0, "Sur": 12, "Villa Arriba Ecológica": 4})
```
